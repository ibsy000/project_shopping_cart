# Build a shopping cart program with the following capabilities:

 1) Takes in an input
 
 2) Stores user input into a dictionary or list
 
 3) The User can add or delete items
 
 4) The User can see current shopping list
 
 5) The program Loops until user 'quits'
 
 6) Upon quiting the program, prints out a receipt of the items with total and quantity.

```python
def shopping(cart = {}):

    while True:

        items = input("Please add an item to your shopping cart: ")
        items = items.strip()

        cart[items] = {}

        # I couldn't figure out how to skip asking for price if the item is already
        # in the cart
        price = input(f"What is the price for {items}? ")
        price = float(price)

        cart[items]['price'] = price


        quantity = input(f"How many {items}s would you like? ")
        quantity = int(quantity)

        cart[items]['quantity'] = quantity

        
        print(f"{quantity} {items}s have been added to your cart.")
            
        
        print("Your current cart:")

        grand_total = []

        
        #This was the best way I could think to show all the added grocery items
        for item in cart.keys():

            print(f"\t{cart[item]['quantity']} {item}s x {cart[item]['price']:.2f}")
            total_price = cart[item]['price'] * cart[item]['quantity']
            grand_total.append(total_price)

        print(f"Total price: {sum(grand_total):.2f}")
        
        
        next = input(f"What would you like to do next? add/delete/show/quit ")
        
        while next not in {'add', 'delete', 'show', 'quit'}:

            next = input("That is not a valid response. add/delete/show/quit ")
            

        while next in {'add', 'delete', 'show', 'quit'}:

        # This was the best way I could think to show the grocery list
            if next == 'show':
                # since this was the best way I could think to show the grocery list
                # I didn't know how to show an empty cart
                print("Your current cart:")

                for item in cart.keys():
                    print(f"\t{cart[item]['quantity']} {item}s x {cart[item]['price']:.2f}")
                    total_price = cart[item]['price'] * cart[item]['quantity']

                print(f"Total price: {sum(grand_total):.2f}")


                next = input(f"What would you like to do next? add/delete/show/quit ")


            if next == 'quit':
                print("thank you for shopping")
                break
        

            if next == 'delete':

                remove = input("Which item would you like to delete? ")

                if remove not in cart.keys():
                    print(f"{remove} is not in your cart")
                    next = input(f"What would you like to do next? add/delete/show/quit ")


                # This was as far as I got.. I couldn't figure out how to delete key:values
                # if the quantity amount got to zero
                elif remove in cart.keys():
                    remove_quantity = input(f"How many of the {cart[remove]['quantity']} {remove}s would you like to delete? ")
                    remove_quantity = int(remove_quantity)

                    new_quantity = cart[remove]['quantity'] - remove_quantity
                    new_quantity = cart[remove]['quantity']

                    remove_price = cart[remove]['price'] * remove_quantity
                    new_total = sum(grand_total) - remove_price


                    print(f"{remove_quantity} {remove}s successfully removed from cart.")
                    next = input(f"What would you like to do next? add/delete/show/quit ")


            if next == 'add':
                shopping()
shopping()
```
